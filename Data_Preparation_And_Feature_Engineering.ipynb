```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts']:
        return 'ศิลปศาสตร์'
    else:
        return name
```

//สร้างฟังก์ชันนี้เพื่อเปลี่ยนชื่อคณะที่เขียนแตกต่างกันให้เป็นชื่อมาตรฐาน เช่น รวมชื่อคณะที่หมายถึง ‘วิทยาศาสตร์’ ให้เป็นคำเดียวกัน และทำเช่นเดียวกันกับ ‘ศิลปศาสตร์’

```python
df['คณะ_มาตรฐาน'] = df['คณะ'].apply(standardize_faculty)
```

//สร้างชุดข้อมูลนักศึกษาที่มีค่าผิดปกติในคอลัมน์ ‘อายุ’ เพื่อใช้ทดสอบการตรวจจับและแก้ไขข้อมูลผิดปกติ

```python
df = pd.DataFrame(data)
```

//แปลงข้อมูลนักศึกษาที่มีอายุผิดปกติเป็น DataFrame เพื่อใช้วิเคราะห์และจัดการข้อมูลต่อไป

```python
print(f"\nค่าเฉลี่ยอายุ (รวมค่าผิดปกติ) - {df['อายุ'].mean():.2f}")
```

//คำนวณและแสดงค่าเฉลี่ยอายุของนักศึกษาทุกคน รวมทั้งค่าผิดปกติ เพื่อดูผลกระทบของ outlier ต่อค่าเฉลี่ย

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
```

//หาค่าควอร์ไทล์ที่ 25% (Q1) และ 75% (Q3) ของอายุ เพื่อใช้ในการตรวจจับค่าผิดปกติโดยใช้ช่วงระหว่างควอร์ไทล์

```python
Q1, Q3
```

//ดูค่าควอร์ไทล์ 25% (Q1) และ 75% (Q3) ของข้อมูลอายุ เพื่อใช้เป็นเกณฑ์ในการตรวจจับค่าผิดปกติหรือ outlier ต่อไป

```python
IQR = Q3 - Q1
```

//คำนวณช่วงระหว่างควอร์ไทล์ (Interquartile Range, IQR) เพื่อใช้กำหนดขอบเขตในการตรวจจับค่าผิดปกติในข้อมูลอายุ

```python
IQR
```

//ดูค่าช่วงระหว่างควอร์ไทล์ (IQR) เพื่อเข้าใจความกระจายของข้อมูลและใช้ในการตรวจจับค่า outlier ต่อไป

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

//คำนวณขอบเขตล่างและบน (lower bound และ upper bound) โดยใช้สูตร IQR เพื่อใช้กรองค่าผิดปกติที่อยู่นอกช่วงนี้

```python
print(lower_bound)
print(upper_bound)
```

//แสดงขอบเขตล่างและบนที่ใช้ในการตรวจจับค่า outlier เพื่อดูช่วงค่าที่ถือว่าเป็นค่าปกติในข้อมูลอายุ

```python
df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

//สร้าง DataFrame ใหม่ที่ตัดค่าอายุผิดปกติ (outlier) ออก เพื่อให้ข้อมูลสะอาดและพร้อมวิเคราะห์

```python
df_no_outliers
```

//ดูข้อมูลหลังกรองเพื่อตรวจสอบว่าแถวที่มีอายุผิดปกติ (เช่น 150) ถูกตัดออกไปแล้ว

```python
df['อายุ'] >= lower_bound
```

//เช็คเงื่อนไขเพื่อดูว่าแต่ละค่าในคอลัมน์ ‘อายุ’ อยู่ในช่วงค่าปกติหรือไม่

```python
df['อายุ'] <= upper_bound
```

//เช็คเงื่อนไขเพื่อดูว่าแต่ละค่าในคอลัมน์ ‘อายุ’ อยู่ในช่วงค่าปกติหรือไม่

```python
data = pd.DataFrame({
    'ชื่อ': ['A', 'B', 'C', 'D'],
    'คณะ': ['วิทย์', 'ศิลป์', 'วิทย์', 'วิศวะ']
})
```

//สร้าง DataFrame ตัวอย่างเพื่อฝึกจัดการกับชื่อคณะที่เขียนต่างกันในข้อมูลขนาดเล็ก

```python
data_encoded = pd.get_dummies(data, columns=['คณะ'])
print(data_encoded)
```

//ใช้คำสั่งนี้เพื่อแปลงคอลัมน์ ‘คณะ’ ให้กลายเป็นคอลัมน์ตัวแปรใหม่แยกตามแต่ละค่า เช่น ‘คณะ_วิทย์’, ‘คณะ_ศิลป์’, ‘คณะ_วิศวะ’ โดยค่าจะเป็น 0 หรือ 1 แสดงว่าค่าในแถวใดเป็นคณะไหน

```python
coffee_data = pd.DataFrame({
    'วันเวลา': pd.to_datetime(['2024-01-15 07:30', '2024-01-15 12:00',
                               '2024-01-15 16:30', '2024-01-16 08:00']),
    'ยอดขาย': [1500, 3200, 2100, 1800]
})
```

//สร้างข้อมูลยอดขายกาแฟ 4 รายการ พร้อมวันและเวลาที่ขาย เพื่อใช้วิเคราะห์ข้อมูลเชิงเวลา

```python
coffee_data
```

//ดูข้อมูลยอดขายกาแฟในแต่ละช่วงเวลาที่บันทึกไว้ เพื่อยืนยันความถูกต้องของข้อมูล

```python
coffee_data['ชั่วโมง'] = coffee_data['วันเวลา'].dt.hour
```

//เพิ่มคอลัมน์ ‘ชั่วโมง’ เพื่อแยกข้อมูลเวลาชั่วโมงจากวันที่-เวลา เพื่อใช้วิเคราะห์ยอดขายตามช่วงเวลา


```python
coffee_data
```

//ดูข้อมูลยอดขายกาแฟพร้อมข้อมูลชั่วโมงที่แยกออกมาเพื่อใช้วิเคราะห์ช่วงเวลาขาย

```python
coffee_data['วันในสัปดาห์'] = coffee_data['วันเวลา'].dt.day_name()
```

//เพิ่มคอลัมน์ ‘วันในสัปดาห์’ เพื่อแยกข้อมูลชื่อวันจากวันที่ เพื่อใช้วิเคราะห์ยอดขายตามวันในสัปดาห์

```python
coffee_data
```

//ดูข้อมูลยอดขายกาแฟพร้อมข้อมูลวันในสัปดาห์ที่แยกออกมา เพื่อวิเคราะห์แนวโน้มตามวัน

```python
coffee_data['ช่วงเวลา'] = pd.cut(coffee_data['ชั่วโมง'],
                                  bins=[0, 10, 14, 18, 24],
                                  labels=['เช้า', 'กลางวัน', 'บ่าย', 'เย็น'])
```

//แบ่งชั่วโมงขายกาแฟออกเป็นช่วงเวลา ‘เช้า’, ‘กลางวัน’, ‘บ่าย’, และ ‘เย็น’ เพื่อวิเคราะห์ยอดขายตามช่วงเวลาของวัน

```python
coffee_data
```

//แสดงข้อมูลยอดขายกาแฟพร้อมรายละเอียดเวลาที่ช่วยให้วิเคราะห์แนวโน้มได้หลากหลายมุมมอง

```python
print(coffee_data[['วันเวลา', 'ช่วงเวลา', 'ยอดขาย']])
```

//ตาราง coffee_data แสดงยอดขายกาแฟพร้อมช่วงเวลา

```python
student_health = pd.DataFrame({
    'ชื่อ': ['สมชาย', 'สมหญิง', 'สมศักดิ์'],
    'ส่วนสูง_cm': [170, 165, 175],
    'น้ำหนัก_kg': [65, 55, 80]
})
```

//สร้างตารางข้อมูลสุขภาพนักเรียน (student_health) ที่ประกอบด้วยชื่อ, ส่วนสูง (ซม.), และน้ำหนัก (กก.) ของแต่ละคน.

```python
student_health['BMI'] = student_health['น้ำหนัก_kg'] / (student_health['ส่วนสูง_cm']/100)**2
student_health['สุขภาพ'] = pd.cut(student_health['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])
```

//คำนวณค่า BMI ของนักศึกษา และจัดกลุ่มสุขภาพเป็น ผอม, ปกติ, อ้วน, หรือ อ้วนมาก ตามช่วงค่า BMI.

```python
student_health
```

//เพื่อคำนวณ BMI และจำแนกสุขภาพเป็นกลุ่ม: ผอม, ปกติ, อ้วน, อ้วนมาก จากส่วนสูงและน้ำหนัก.

```python
print(student_health[['ชื่อ', 'BMI', 'สุขภาพ']])
```

//แสดงชื่อ, ค่าดัชนีมวลกาย (BMI), และสถานะสุขภาพของแต่ละคน.

```python
raw_data = {
    'รหัสนักศึกษา': ['64001', '64002', '64003', '64004', '64005'],
    'ชื่อ': ['สมชาย', 'สมหญิง', None, 'สมศักดิ์', 'สมใจ'],
    'คณะ': ['วิทย์', 'ศิลปศาสตร์', 'วิทยาศาสตร์', 'Arts', 'วิศวะ'],
    'ชั้นปี': [2, 3, 2, None, 4],
    'GPA': [3.2, 3.8, -1, 2.9, 4.5],  # -1 และ 4.5 ผิดปกติ
    'กิจกรรม': ['กีฬา', 'ดนตรี', 'กีฬา,ดนตรี', None, 'อาสา'],
    'เวลาเข้าร่วม': ['08:00', '9.30', '08:30', '08:00', None]
}
df_student = pd.DataFrame(raw_data)
```

//สร้างตารางข้อมูลนักศึกษาที่มีทั้งข้อมูลครบและข้อมูลผิดปกติ/หาย

```python
faculty_mapping = {
    'วิทย์': 'วิทยาศาสตร์',
    'วิทยาศาสตร์': 'วิทยาศาสตร์',
    'ศิลปศาสตร์': 'ศิลปศาสตร์',
    'Arts': 'ศิลปศาสตร์',
    'วิศวะ': 'วิศวกรรมศาสตร์'
}
```

//แผนที่นี้ใช้แมปชื่อคณะต่างๆ

```python
faculty_mapping
```

//ใช้แปลงชื่อคณะให้เป็นมาตรฐานเดียวกัน เพื่อความสอดคล้องของข้อมูล

```python
df_student['คณะ'] = df_student['คณะ'].map(faculty_mapping)
```

//แก้ไขชื่อคณะในข้อมูลนักศึกษาให้ตรงตามมาตรฐานที่กำหนด เพื่อให้ข้อมูลสะอาดและสอดคล้องกันมากขึ้น

```python
df_student
```
